```python
import pandas as pd
import numpy as np

missing_values = ["n/a", "na", "--","?",'None']

chefmozaccepts=pd.read_csv('chefmozaccepts.csv',na_values = missing_values)
chefmozcuisine=pd.read_csv('chefmozcuisine.csv',na_values = missing_values)
chefmozhours4=pd.read_csv('chefmozhours4.csv',na_values = missing_values)
chefmozparking=pd.read_csv('chefmozparking.csv',na_values = missing_values)
rating_final=pd.read_csv('rating_final.csv',na_values = missing_values)
usercuisine=pd.read_csv('usercuisine.csv',na_values = missing_values)
userpayment=pd.read_csv('userpayment.csv',na_values = missing_values)
userprofile=pd.read_csv('userprofile.csv',na_values = missing_values)
geoplaces2=pd.read_csv('geoplaces2.csv',encoding='latin-1',na_values = missing_values)

geoplaces2['latitude','']
```

## 2.1.2

```python
# check for any duplicated row based on all columns and drop if any

if chefmozaccepts.duplicated().sum()==0:
        print(" No duplicates found in this file")
        print()
else:              
        print(" There are duplicates in file name ",'chefmozaccepts')
        print()
        chefmozaccepts.drop_duplicates(inplace=True) 
if chefmozcuisine.duplicated().sum()==0:
        print(" No duplicates found in this file")
        print()
else:   
        print(" There are duplicates in file name ",'chefmozcuisine')         
        print()
        chefmozcuisine.drop_duplicates(inplace=True) 
if chefmozhours4.duplicated().sum()==0:
        print(" No duplicates found in this file")
        print()
else:              
        print(" There are duplicates in file name ",'chefmozhours4')
        print()
        chefmozhours4.drop_duplicates(inplace=True) 

if chefmozparking.duplicated().sum()==0:
        print(" No duplicates found in this file")
        print()
else:              
        print(" There are duplicates in file name ",'chefmozparking')
        print()
        chefmozparking.drop_duplicates(inplace=True) 

if rating_final.duplicated().sum()==0:
        print(" No duplicates found in this file")
        print()
else:              
        print(" There are duplicates in file name ",' rating_final')
        print()
        rating_final.drop_duplicates(inplace=True) 

if usercuisine.duplicated().sum()==0:
        print(" No duplicates found in this file")
        print()
else:              
        print(" There are duplicates in file name ",'usercuisine')
        print()
        usercuisine.drop_duplicates(inplace=True) 
if userpayment.duplicated().sum()==0:
        print(" No duplicates found in this file")
        print()
else:              
        print(" There are duplicates in file name ",'userpayment')
        print()
        userpayment.drop_duplicates(inplace=True) 

if userprofile.duplicated().sum()==0:
        print(" No duplicates found in this file")
        print()
else:              
        print(" There are duplicates in file name ",'userprofile')
        print()
        userprofile.drop_duplicates(inplace=True) 

if geoplaces2.duplicated().sum()==0:
        print(" No duplicates found in this file")
        print()
else:              
        print(" There are duplicates in file name ",'geoplaces2')
        print()
        geoplaces2.drop_duplicates(inplace=True) 

# If there is a row or column where all or most values are missing, drop them.

userprofile.dropna(thresh=(0.4*len(userprofile.index)),axis=1,inplace=True)
geoplaces2.dropna(thresh=(0.6*len(geoplaces2.index)),axis=1,inplace=True) # columns 'fax' , 'zip' and 'url' arre droped in this step since they have more than 60% missing values

#Drop (‘fax’,’zip’,’url’) columns from ‘geoplaces2.csv’.
if set(['fax','zip','url']).issubset(geoplaces2.columns):
    geoplaces2.drop(columns=['fax', 'zip','url'])
    print("in to drop.......")
    
#Find the missing values (i.e. ‘?’,’None’, etc.). First replace them with nan in Numpy and then fill in them with appropriate values (methods).

print(" Missing values in the file name :",'chefmozaccepts' )
print(chefmozaccepts.isnull().sum())
print()

print(" Missing values in the file name :",'chefmozcuisine' )
print(chefmozcuisine.isnull().sum())
print()

print(" Missing values in the file name :",'chefmozhours4' )
print(chefmozhours4.isnull().sum())
print()

print(" Missing values in the file name :",'chefmozparking' )
print(chefmozparking.isnull().sum())
print()

print(" Missing values in the file name :",'rating_final' )
print(rating_final.isnull().sum())
print()

print(" Missing values in the file name :",'usercuisine' )
print(usercuisine.isnull().sum())
print()

print(" Missing values in the file name :",'userpayment' )
print(userpayment.isnull().sum())
print()

print(" Missing values in the file name :",'userprofile' )
print(userprofile.isnull().sum())
print()

print(" Missing values in the file name :",'geoplaces2' )
print(geoplaces2.isnull().sum())
print()
        
# From the above code we can see that there are missing values in the files 'userprofile' and 'geoplaces2'  hence they must be filled with appropriet values      
# filling the  missing values in the smoker column
val=userprofile.loc[userprofile['userID'].isin(userprofile[userprofile['smoker'].isnull()]['userID'])]
for index, row in val.head().iterrows():         
        userprofile.at[index, 'smoker'] = lambda row: 'false' if row['drink_level']=='abstemious' else 'true'
# Filling the missing values in the  column 'dress_preference'
userprofile['dress_preference'].bfill(axis ='rows',inplace=True)
        
# Filling the missing values in the  column 'activity'   
userprofile['activity'].bfill(axis ='rows',inplace=True)
#userprofile['ambience'].bfill(axis ='rows',inplace=True)
userprofile['budget'].bfill(axis ='rows',inplace=True)
# Filling the missing values in the column 'ambience'
amb=userprofile.loc[userprofile['userID'].isin(userprofile[userprofile['ambience'].isnull()]['userID'])]
for index, row in amb.head().iterrows(): 
    userprofile.at[index, 'ambience'] = lambda row: 'family' if row['drink_level']=='abstemious' else 'friends'
# Filling the missing values in the  column 'transport'
trans=userprofile.loc[userprofile['userID'].isin(userprofile[userprofile['transport'].isnull()]['userID'])]
for index, row in trans.head().iterrows():          
          userprofile.at[index,'transport'] = lambda row: 'car owner' if row['ambience']=='family' else 'on foot'
# Filling the missing values in the  column 'marital_status'
status=userprofile.loc[userprofile['userID'].isin(userprofile[userprofile['marital_status'].isnull()]['userID'])]
for index, row in status.head().iterrows(): 
     userprofile.at[index,'marital_status'] =lambda row: "married" if row['transport']=='car owner' else 'single' 
# Filling the missing values in the  column 'hijos'
sta=userprofile.loc[userprofile['userID'].isin(userprofile[userprofile['hijos'].isnull()]['userID'])]
for index, row in status.head().iterrows():  
     userprofile.at[index,'hijos'] = lambda row: "kids" if row['marital_status']=='married' else 'independent'                    
# Fill the remaining cells which has not been filled from the above process
userprofile.fillna(method='ffill',inplace=True)
# Filling the missing values in  the file geoplaces2

geoplaces2.fillna(method='ffill',inplace=True)



print('-------------------------')
print(userprofile.isnull().sum())    

print('-------------------------')
print(geoplaces2.isnull().sum())   
```
